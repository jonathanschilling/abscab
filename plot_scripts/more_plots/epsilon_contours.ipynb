```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nr = 100
nz = 100

rpGrid = np.linspace(-1.0, 2.0, nr)
zpGrid = np.linspace(-0.52, 1.51, nz)

rp, zp = np.meshgrid(rpGrid, zpGrid)

def r_i(rp, zp):
    return np.hypot(rp, zp)

def r_f(rp, zp):
    return np.hypot(rp, 1.0 - zp)

def eps(ri, rf):
    return 1.0 / (ri + rf)

ri = r_i(rp, zp)
rf = r_f(rp, zp)

e = eps(ri, rf)

figw = 5.0 # inches
plt.figure(figsize=(figw,3.5/6*figw))
plt.contourf(rpGrid, zpGrid, e)
plt.axhline(0.0, color='k', linewidth=0.5)
plt.axhline(1.0, color='k', linewidth=0.5)
plt.axvline(0.0, color='k', linewidth=0.5)
plt.plot([0.0, 0.0], [0.0, 1.0], 'r-', linewidth=2.0)
cb = plt.colorbar()
cb.set_label(r"$\epsilon$")
plt.axis("equal")
plt.xlabel(r"$\rho'$")
plt.ylabel(r"$z'$")

#plt.tight_layout()
plt.subplots_adjust(left=0.15,
                    right=0.99,
                    top = 0.95,
                    bottom = 0.17)

plt.savefig("epsilon_contours.png")
plt.savefig("epsilon_contours.eps")

plt.savefig("../article/img/epsilon_contours.eps")
```
